```python
from google.colab import files
from io import BytesIO
from PIL import Image
```

```python
uploaded = files.upload()
im = Image.open(BytesIO(uploaded['good_0.jpg']))

img1 = cv2.imread("good_0.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(img1)

equ1 = cv2.equalizeHist(img1)
cv2_imshow(equ1)
```

```python
f = np.fft.fft2(equ1)
fshift = np.fft.fftshift(f)
ms = 20*np.log(np.abs(fshift))
#ms = 20*np.log(np.abs(f))
#print(ms)
ms1 = np.asarray(ms, dtype = np.uint8)
cv2_imshow(ms1)
```

```python
histg = cv2.calcHist([ms1],[0],None,[256],[0,256]) 
plt.plot(histg)
plt.show()
```

```python
f, axarr = plt.subplots(5,4)
for i in range(5):
  im = Image.open(BytesIO(uploaded['colour'+str(i)+'.jpg']))
  print("grey scale image :",i+1)
  im = cv2.imread("colour"+str(i)+".jpg", cv2.IMREAD_GRAYSCALE)
 
  axarr[i,0]=cv2_imshow(im)
  print(" equalized grey scale image 1:")
  equco = cv2.equalizeHist(im)
  axarr[i,1]=cv2_imshow(equco)
  print("Image after erosion:")
  img_erosion1 = cv2.erode(equco,kernel, iterations=1)
  axarr[i,2]=img_erosion1
  cv2_imshow(img_erosion1)
  print("Image after dilation:")
  img_dilation1 = cv2.dilate(equco, kernel, iterations=1)
  axarr[i,3]=  img_dilation1
  cv2_imshow(img_dilation1)
  print("Binary image of erosion:")
  (thresh, th) = cv2.threshold(img_erosion1, 127, 255, cv2.THRESH_BINARY)
  cv2_imshow(th)
f.subplots_adjust(hspace=0.3)
plt.show()
```
